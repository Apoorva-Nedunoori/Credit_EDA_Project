Clearly there are outliers in all the AMT_variables, so removing the outliers now.. 


```python
inp1_outliers = ['AMT_INCOME_TOTAL','AMT_CREDIT','AMT_ANNUITY','AMT_GOODS_PRICE', 'DAYS_EMPLOYED', 'DAYS_BIRTH']
new_inp1 = inp1[inp1_outliers]

for i in new_inp1.columns:
    Q1 = new_inp1[i].quantile(0.25)
    Q3 = new_inp1[i].quantile(0.75)

    IQR = Q3 - Q1
    
    
    lower_fence = Q1 - 1.5*IQR
    upper_fence = Q3 + 1.5*IQR
    
    new_inp1[(new_inp1.AMT_CREDIT <= upper_fence)].head()
    new_inp1[i][new_inp1[i] <= lower_fence] = lower_fence
    new_inp1[i][new_inp1[i] >= upper_fence] = upper_fence
    
    print("OUTLIERS:",i,lower_fence,upper_fence)
```

###### COMPARATIVE ANALYSIS of CATEGORICAL VARIABLE (AMT_CREDIT) before and after cleaning the outliers in inp1

```python
#With outliers
plt.figure(1, figsize = (10,5))
plt.subplot(121)
boxplot = inp1.boxplot(['AMT_CREDIT'], notch=True, patch_artist=True)
boxplot.title.set_text('With Outliers')
boxplot.set(facecolor = 'turquoise' )

#Without Outliers
plt.subplot(122)
boxplot = new_inp1.boxplot(['AMT_CREDIT'], notch=True, patch_artist=True)
boxplot.title.set_text('Without Outliers')
boxplot.set(facecolor = 'gold' )
plt.show()

min = inp1['AMT_CREDIT'].describe().min()
max = inp1['AMT_CREDIT'].describe().max()
range = [min, max]
print("Before removing:",range)

min = new_inp1['AMT_CREDIT'].describe().min()
max = new_inp1['AMT_CREDIT'].describe().max()
range = [min, max]
print("After removing:",range)

inp1['AMT_CREDIT'].describe()
```
